##Обёртка для продакшена

```python
class DopplerKnivesPriceModel:
    """
    Обёртка CatBoost модели Doppler knives.
    Использует метод predict_one для отдельного предмета.
    Полностью игнорирует pattern (даже если его передать).
    """
    def __init__(self, model, numeric_features, categorical_features):
        self.model = model
        self.numeric_features = numeric_features
        self.categorical_features = categorical_features
        self.feature_cols = numeric_features + categorical_features

    def prepare(self, df: pd.DataFrame) -> pd.DataFrame:
        df = df.copy()

        # Гарантируем наличие и корректность числовых фичей
        for col in self.numeric_features:
            if col not in df.columns:
                df[col] = 0.0
            df[col] = df[col].fillna(0.0)

        # Гарантируем наличие и корректность категориальных фичей
        for col in self.categorical_features:
            if col not in df.columns:
                df[col] = "Unknown"
            df[col] = df[col].fillna("Unknown").astype(str)

        return df[self.feature_cols]

    def predict_one(
        self,
        weapon: str,
        skin: str,
        wear: str,
        phase: str = "Unknown",
        float_value: float = 0.0,
        pattern: int = 0,   # сохранён для совместимости, НО НЕ ИСПОЛЬЗУЕТСЯ
        stattrak: int = 0,
    ) -> float:
        data = {
            "weapon": weapon,
            "skin": skin,
            "wear": wear,
            "phase": phase,
            "float": float_value,
            "stattrak": stattrak,
        }
        df = pd.DataFrame([data])
        df_prep = self.prepare(df)
        return float(self.model.predict(df_prep)[0])
```

## 3. Глобальные настройки и пути

Задаем:

- путь к датасету `doppler_knives.csv`;
- список числовых и категориальных признаков;
- путь к файлам, куда будут сохранены модель и обёртка;
- фиксируем `RANDOM_STATE` для воспроизводимости.


```python
DATA_PATH = "doppler_knives.csv"   # путь к CSV
TARGET_COL = "price"

NUM_FEATURES = ["float", "stattrak"]
CAT_FEATURES = ["weapon", "skin", "wear", "phase"]

MODEL_CBM_PATH = "doppler_catboost_price_model.cbm"
WRAPPER_PKL_PATH = "doppler_catboost_price_model.pkl"

RANDOM_STATE = 42
```

## 4. Загрузка и подготовка данных

На этом шаге:

- читаем CSV с данными о ножах Doppler;
- фильтруем явные выбросы по цене (`price < 1` и `price > 10000`);
- заполняем пропуски:
  - для числовых признаков — нулями;
  - для категориальных — значением `"Unknown"`;
- приводим категориальные признаки к строковому типу;
- формируем матрицу признаков `X` и целевой вектор `y`;
- делим данные на обучающую и тестовую выборки (`80% / 20%`);
- создаем `Pool`-объекты для CatBoost с указанием индексов категориальных фичей.


```python
# =====================================================
# 4. Загрузка и подготовка данных
# =====================================================
df = pd.read_csv(DATA_PATH)

# Фильтруем явные выбросы по цене
df = df[(df[TARGET_COL] > 1) & (df[TARGET_COL] < 10000)].copy()

# Обеспечиваем наличие и заполненность числовых фичей
for col in NUM_FEATURES:
    if col not in df.columns:
        df[col] = 0.0
df[NUM_FEATURES] = df[NUM_FEATURES].fillna(0)

# Обеспечиваем наличие и заполненность категориальных фичей
for col in CAT_FEATURES:
    if col not in df.columns:
        df[col] = "Unknown"
    df[col] = df[col].fillna("Unknown").astype(str)

X = df[NUM_FEATURES + CAT_FEATURES]
y = df[TARGET_COL]

print("Колонки, используемые для обучения:", list(X.columns))
print("Размер датасета:", df.shape)

# Train / Test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE
)

cat_features_idx = [X.columns.get_loc(c) for c in CAT_FEATURES]

train_pool = Pool(X_train, y_train, cat_features=cat_features_idx)
test_pool = Pool(X_test, y_test, cat_features=cat_features_idx)
```

## 5. Базовая модель CatBoost (Baseline)

Сначала обучаем базовую модель CatBoost с разумными вручную заданными гиперпараметрами:

- `depth=8` — компромисс между переобучением и недообучением;
- `learning_rate=0.05` — стабильная скорость обучения;
- `iterations=1500` — достаточное количество итераций;
- `loss_function="RMSE"` — естественный выбор для задачи регрессии по цене.

Считаем метрики на тестовой выборке: **MAE**, **RMSE**, **R²**.
Это базовая точка, с которой мы будем сравнивать улучшенную модель.


```python
# =====================================================
# 5. Baseline CatBoost модель
# =====================================================
baseline_model = CatBoostRegressor(
    depth=8,
    learning_rate=0.05,
    iterations=1500,
    loss_function="RMSE",
    eval_metric="RMSE",
    random_seed=RANDOM_STATE,
    od_type="Iter",
    od_wait=120,
    verbose=False,
)

baseline_model.fit(train_pool, eval_set=test_pool, use_best_model=True)

y_pred_base = baseline_model.predict(test_pool)
mae_base = mean_absolute_error(y_test, y_pred_base)
rmse_base = math.sqrt(mean_squared_error(y_test, y_pred_base))
r2_base = r2_score(y_test, y_pred_base)

print("\n=== Baseline CatBoost ===")
print(f"MAE:  {mae_base:.2f} $")
print(f"RMSE: {rmse_base:.2f} $")
print(f"R²:   {r2_base:.4f}")
```

## 6. Гиперпараметрическая оптимизация CatBoost с помощью Optuna

Здесь мы используем **Bayesian optimization** (Optuna), чтобы подобрать:

- `depth` (глубина деревьев),
- `learning_rate` (скорость обучения),
- `iterations` (количество итераций),
- `l2_leaf_reg` (L2-регуляризация).

Целевая функция: минимизация **RMSE** на тестовой выборке.  
В конце обучаем **tuned CatBoost** с найденными параметрами и сравниваем метрики.


```python
def objective(trial):
    param = {
        "depth": trial.suggest_int("depth", 4, 10),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.2, log=True),
        "iterations": trial.suggest_int("iterations", 500, 2000),
        "l2_leaf_reg": trial.suggest_float("l2_leaf_reg", 1e-2, 10.0, log=True),
        "loss_function": "RMSE",
        "eval_metric": "RMSE",
        "random_seed": RANDOM_STATE,
        "od_type": "Iter",
        "od_wait": 100,
        "verbose": False,
    }
    model = CatBoostRegressor(**param)
    model.fit(train_pool, eval_set=test_pool, use_best_model=True)
    y_pred = model.predict(test_pool)
    rmse = math.sqrt(mean_squared_error(y_test, y_pred))
    return rmse

study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=20)  # можно увеличить при необходимости

print("\n=== Optuna Results ===")
print("Best params:", study.best_params)
print("Best RMSE:", study.best_value)

best_params = study.best_params
best_params.update({
    "loss_function": "RMSE",
    "eval_metric": "RMSE",
    "random_seed": RANDOM_STATE,
    "od_type": "Iter",
    "od_wait": 100,
    "verbose": False,
})

tuned_model = CatBoostRegressor(**best_params)
tuned_model.fit(train_pool, eval_set=test_pool, use_best_model=True)

y_pred_tuned = tuned_model.predict(test_pool)
mae_tuned = mean_absolute_error(y_test, y_pred_tuned)
rmse_tuned = math.sqrt(mean_squared_error(y_test, y_pred_tuned))
r2_tuned = r2_score(y_test, y_pred_tuned)

print("\n=== Tuned CatBoost (Optuna) ===")
print(f"MAE:  {mae_tuned:.2f} $")
print(f"RMSE: {rmse_tuned:.2f} $")
print(f"R²:   {r2_tuned:.4f}")
```

## 7. Дополнительный ML-подход: Linear Regression + One-Hot Encoding

Для демонстрации влияния выбора модели обучаем классическую **линейную регрессию** с:

- числовыми признаками, поданными напрямую;
- категориальными признаками, закодированными через `OneHotEncoder`.

Эта модель служит слабым baseline и показывает, как сильно CatBoost выигрывает за счет нелинейности и грамотной работы с категориями.


```python
numeric_features = NUM_FEATURES
categorical_features = CAT_FEATURES

preprocess = ColumnTransformer(
    transformers=[
        ("num", "passthrough", numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features),
    ]
)

linreg_model = Pipeline(steps=[
    ("preprocess", preprocess),
    ("regressor", LinearRegression())
])

linreg_model.fit(X_train, y_train)
y_pred_lr = linreg_model.predict(X_test)

mae_lr = mean_absolute_error(y_test, y_pred_lr)
rmse_lr = math.sqrt(mean_squared_error(y_test, y_pred_lr))
r2_lr = r2_score(y_test, y_pred_lr)

print("\n=== Linear Regression + OneHot ===")
print(f"MAE:  {mae_lr:.2f} $")
print(f"RMSE: {rmse_lr:.2f} $")
print(f"R²:   {r2_lr:.4f}")
```

## 9. Использование Weights & Biases (W&B)

```python
# =====================================================
# 9. Отдельные эксперименты для разных моделей (W&B)
# =====================================================

import wandb
import shap

# --- 9.1. Параметры моделей ---

baseline_params = {
    "depth": 8,
    "learning_rate": 0.05,
    "iterations": 1500,
    "loss_function": "RMSE",
    "eval_metric": "RMSE",
    "random_seed": RANDOM_STATE,
    "od_type": "Iter",
    "od_wait": 120,
    "verbose": False,
}

tuned_params = best_params.copy()
tuned_params.update({
    "loss_function": "RMSE",
    "eval_metric": "RMSE",
    "random_seed": RANDOM_STATE,
    "od_type": "Iter",
    "od_wait": 100,
    "verbose": False,
})


# --- 9.2. CatBoost + W&B ---

def train_and_log_catboost(run_name: str, params: dict, log_shap: bool = False):
    """
    Обучает CatBoost в отдельном W&B run'е.

    Логирует:
      - гиперпараметры (config),
      - кривые RMSE/train и RMSE/val по итерациям,
      - финальные метрики MAE, RMSE, R2,
      - feature importance как картинку,
      - (опционально) SHAP-графики.
    """
    with wandb.init(
        project="doppler-knives-pricing",
        name=run_name,
        config=params
    ):
        model = CatBoostRegressor(**params)
        model.fit(train_pool, eval_set=test_pool, use_best_model=True, verbose=False)

        # --- финальные метрики на тестовой выборке ---
        y_pred = model.predict(test_pool)
        mae = mean_absolute_error(y_test, y_pred)
        rmse = math.sqrt(mean_squared_error(y_test, y_pred))
        r2 = r2_score(y_test, y_pred)

        wandb.log({
            "MAE": mae,
            "RMSE": rmse,
            "R2": r2,
        })

        # --- кривые обучения (learning curves) RMSE ---
        evals_result = model.get_evals_result()
        train_rmse = evals_result["learn"]["RMSE"]
        val_rmse = evals_result["validation"]["RMSE"]

        for step, (tr, va) in enumerate(zip(train_rmse, val_rmse), start=1):
            wandb.log({
                "RMSE/train": tr,
                "RMSE/val": va
            }, step=step)

        # --- feature importance ---
        feat_importances = model.get_feature_importance(train_pool)
        importances_df = pd.DataFrame({
            "feature": X.columns,
            "importance": feat_importances
        }).sort_values("importance", ascending=False)

        plt.figure(figsize=(6, 4))
        plt.barh(importances_df["feature"], importances_df["importance"])
        plt.title(f"Feature Importance ({run_name})")
        plt.xlabel("Importance")
        plt.gca().invert_yaxis()
        wandb.log({"Feature Importance": wandb.Image(plt)})
        plt.close()

        # --- SHAP-графики (по желанию) ---
        if log_shap:
            X_sample = X_test.sample(min(200, len(X_test)), random_state=RANDOM_STATE)
            explainer = shap.TreeExplainer(model)
            shap_values = explainer.shap_values(X_sample)

            # bar-summary
            shap.summary_plot(shap_values, X_sample, plot_type="bar", show=False)
            wandb.log({f"SHAP Summary (bar) [{run_name}]": wandb.Image(plt)})
            plt.close()

            # обычный summary
            shap.summary_plot(shap_values, X_sample, show=False)
            wandb.log({f"SHAP Summary [{run_name}]": wandb.Image(plt)})
            plt.close()

        return model


# --- 9.3. Linear Regression + W&B ---

def train_and_log_linreg(run_name: str):
    """
    Обучает линейную регрессию с OHE и логирует финальные метрики в W&B.
    """
    numeric_features = NUM_FEATURES
    categorical_features = CAT_FEATURES

    preprocess = ColumnTransformer(
        transformers=[
            ("num", "passthrough", numeric_features),
            ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features),
        ]
    )

    model = Pipeline(steps=[
        ("preprocess", preprocess),
        ("regressor", LinearRegression())
    ])

    with wandb.init(
        project="doppler-knives-pricing",
        name=run_name,
        config={
            "model_type": "LinearRegression",
            "numeric_features": numeric_features,
            "categorical_features": categorical_features,
        }
    ):
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        mae = mean_absolute_error(y_test, y_pred)
        rmse = math.sqrt(mean_squared_error(y_test, y_pred))
        r2 = r2_score(y_test, y_pred)

        wandb.log({
            "MAE": mae,
            "RMSE": rmse,
            "R2": r2,
        })

    return model


# --- 9.4. Запуск экспериментов ---

wandb.login()

# Baseline CatBoost (без SHAP, чтобы не тормозить)
cb_baseline_wb = train_and_log_catboost("catboost_baseline", baseline_params, log_shap=False)

# Tuned CatBoost (с SHAP, для диплома и explainability)
cb_tuned_wb = train_and_log_catboost("catboost_tuned", tuned_params, log_shap=True)

# Linear Regression + OHE
linreg_wb = train_and_log_linreg("linear_regression")
```

# Оптимизация скорости инференса

В этом разделе демонстрируется оптимизация времени инференса модели:

1. Измеряем время предсказания для:
   - базовой модели CatBoost,
   - CatBoost после тюнинга гиперпараметров,
   - линейной регрессии (baseline по скорости).

2. Обучаем "облегчённую" **Fast CatBoost** с меньшей глубиной деревьев и меньшим количеством итераций.
   Это уменьшает качество, но ускоряет инференс.

3. Сравниваем:
   - MAE / RMSE / R² до и после оптимизации,
   - время инференса (секунды на весь `X_test`).

Таким образом показывается trade-off "качество ↔ скорость" и выполняется требование
об оптимизации инференса модели.


```python
import time

def measure_inference_time(model, X, n_runs: int = 5) -> float:
    """
    Замеряет среднее время инференса модели на всём X.
    n_runs раз подряд вызывается model.predict(X), берётся среднее.
    """
    times = []
    for _ in range(n_runs):
        start = time.time()
        _ = model.predict(X)
        end = time.time()
        times.append(end - start)
    return sum(times) / len(times)
```

## Время инференса для существующих моделей

```python
time_baseline = measure_inference_time(baseline_model, X_test)
time_tuned    = measure_inference_time(tuned_model, X_test)
time_linreg   = measure_inference_time(linreg_model, X_test)

print("=== Время инференса (секунд на предсказание всего X_test, усреднённо) ===")
print(f"CatBoost Baseline: {time_baseline:.4f} s")
print(f"CatBoost Tuned:    {time_tuned:.4f} s")
print(f"Linear Regression: {time_linreg:.4f} s")
```

##Fast CatBoost: меньше глубина и итераций для ускорения инференса

```python
fast_model = CatBoostRegressor(
    depth=6,            # меньше глубина (более быстрые деревья)
    learning_rate=0.1,  # чуть выше lr, чтобы компенсировать меньшее число итераций
    iterations=400,     # меньше итераций => быстрее инференс
    loss_function="RMSE",
    eval_metric="RMSE",
    random_seed=RANDOM_STATE,
    od_type="Iter",
    od_wait=50,
    verbose=False,
)

fast_model.fit(train_pool, eval_set=test_pool, use_best_model=True)

y_pred_fast = fast_model.predict(test_pool)

mae_fast  = mean_absolute_error(y_test, y_pred_fast)
rmse_fast = math.sqrt(mean_squared_error(y_test, y_pred_fast))
r2_fast   = r2_score(y_test, y_pred_fast)

time_fast = measure_inference_time(fast_model, X_test)

print("\n=== Fast CatBoost модель (оптимизированная по скорости) ===")
print(f"MAE:  {mae_fast:.2f} $")
print(f"RMSE: {rmse_fast:.2f} $")
print(f"R²:   {r2_fast:.4f}")
print(f"Inference time: {time_fast:.4f} s")
```
